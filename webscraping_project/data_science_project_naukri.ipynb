FETCHING DETAILS LIKE JOB NAME, LOCATION, SKILLS NEEDED, ORGANIZATION NAME FOR MULTIPLE CITIES

FIRST CITY: BANGALORE

```python
jobs_bangalore = []
skills_bangalore = []
company_bangalore = []
city_bangalore = []
```

```python
base_url = 'https://www.naukri.com/data-scientist-jobs-in-'
loc = 'bangalore'
page = 1

while page < 15:
    #print (page)
    url = base_url + loc + "-" + str(page)
    url_source = requests.get(url)
    url_soup = soup(url_source.text, 'html.parser')
    content = url_soup.findAll('a', {'class': 'content'})
    for name in content:
                
        job_name = name.findAll('li',{'class': 'desig'})[0].text
        #print(job_name)
        jobs_bangalore.append(job_name)
                
        skills = name.findAll('span',{'class': 'skill'})[0].text
        #print(skills)
        skills_bangalore.append(skills)
                
        industry = name.findAll('span',{'class': 'org'})[0].text
        #print(industry)
        company_bangalore.append(industry)        
                
        location = name.findAll('span',{'class': 'loc'})[0].text
        #print(location)
        city_bangalore.append(location)
        
    page+=1
```

```python
print(len(jobs_bangalore))
print(len(skills_bangalore))
print(len(company_bangalore))
print(len(city_bangalore))
```

```python
# Removing the spaces between the skills and treating multiple string in a skill as a single one for analysis
import re

skill_set = []
set_of_skills = []

for i in range (0,1553):
    p = re.sub("[^a-zA-Z,]", "", skills[i])
    p = p.lower()
    p = p.replace("   ","")
    q = p.split(",")
    #print(p)
    set_of_skills.append(p)
    skill_set.append(q)
    #print (skill_set[i])
#print (i)

print("The actual skills list was: ", skill_set[-1])

print("\n The modified skills list is:", set_of_skills[-1])
```

```python
# Alloting both the skills list to the dataframe

dataset["Skillset"] = set_of_skills
dataset["Skillset_list"] = skill_set
dataset.head()
```

SPLITTING DOWN THE SKILLS AS VARIOUS CATEGORIES FOR BETTER INSIGHT

```python
prog_lang_dict = Counter({'R':cnt['r'],'Python':cnt['python'],'Java':cnt['java'], 'Ruby':cnt['ruby'], 'Perl':cnt['perl'], 
                          'Matlab':cnt['matlab'], 'Mathematica':cnt['mathematica'], 'Php':cnt['php'],
                          'JavaScript':cnt['javascript'], 'Scala': cnt['scala'], 'C':cnt['c']})
                      
analysis_tool_dict = Counter({'Excel':cnt['excel'],  'Tableau':cnt['tableau'],'SAS':cnt['sas'],'SPSS':cnt['spss'],
                              'Spotfire': cnt['spotfire']})

hadoop_dict = Counter({'Hadoop':cnt['hadoop'], 'MapReduce':cnt['mapreduce'],'Spark':cnt['spark'], 'Pig':cnt['pig'],
                       'Hive':cnt['hive'],'Oozie':cnt['oozie'], 'Flume':cnt['flume'], 'Algorithms':cnt['algorithms']})
    
analytics_dict = Counter({'Data Analysis':cnt['dataanalysis'],'Data Mining':cnt['datamining'],'Analytics':cnt['analytics'],
                          'Business Analytics':cnt['businessanalytics'], 'Business Analysis':cnt['businessanalysis'],
                          'Azure':cnt['azure'], 'AWS':cnt['aws'], 'Data Science':cnt['datascience'],
                          'Business Analyst':cnt['businessanalyst'], 'Data Scientist':cnt['datascientist'],
                          'Analysis':cnt['analytical'], 'BigData':cnt['bigdata']})
                
database_dict = Counter({'SQL':cnt['sql'], 'NoSQL':cnt['nosql'],'HBase':cnt['hbase'], 'Cassandra':cnt['cassandra'],
                         'Tools':cnt['tools']})
                    
# About Education         
    
edu_dict = Counter({'Bachelor':cnt['bachelor'],'Master':cnt['master'],'PhD': cnt['phd'],'MBA':cnt['mba']})
                          
# Calculate requested languages other than english
lang_dict = Counter({'French':cnt['french'],'German':cnt['german'],'Spanish':cnt['spanish'],'Chinese':cnt['chinese'],
                     'Japanese':cnt['japanese']})
          
# Education counters 2 Useless for analitycs but usefull for model
course_dict = Counter({'Computer Science':cnt['computerscience'],'Statistics':cnt['statistics'], 
                          'Mathematics':cnt['mathematics'],'Physics':cnt['physics'], 'Machine Learning':cnt['machinelearning'], 
                          'Economics':cnt['economics'], 'Software Engineer': cnt['softwareengineer'],'IT':cnt['it']})

top_10_skills = Counter({'R':cnt['r'],'Python':cnt['python'],'Machine Learning':cnt['machinelearning'],
                         'SAS':cnt['sas'],'SQL':cnt['sql'], 'Data Analysis':cnt['dataanalysis'],'Analytics':cnt['analytical'],
                         'Data Scientist':cnt['datascientist'],'Data Mining':cnt['datamining'],'Algorithms':cnt['algorithms'], 
                         'Tools':cnt['tools'], 'Business Analysis':cnt['businessanalysis']})

# Combine our Counter objects

main_skills = top_10_skills
skl_dict =  edu_dict + lang_dict +  course_dict
overall_total_skills = prog_lang_dict + analysis_tool_dict + hadoop_dict + database_dict  + analytics_dict + skl_dict

print ("prog_lang_dict:\n", prog_lang_dict,"\n\nanalysis_tool_dict:\n", analysis_tool_dict,"\n\nhadoop_dict:\n", hadoop_dict,
       "\n\nanalytics_dict:\n", analytics_dict,"\n\ndatabase_dict:\n", database_dict,"\n\ncourse_dict:\n", course_dict,
       "\n\nlang_dict:\n", lang_dict,"\n\nedu_dict:\n", edu_dict)
print ("\n\n Overall Key Skills summary:\n", overall_total_skills )
print ("\n\n Main Skills needed:\n", main_skills )
```
